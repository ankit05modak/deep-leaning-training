```python
def print_unique_col_values(df):
    for column in df:
            if df[column].dtype == "object":
                print(f"{column} : {df1[column].unique()}")
                
print_unique_col_values(df1)
```

```python
df1.replace("No internet service", "No", inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
df1.replace("No phone service", "No", inplace=True)
print_unique_col_values(df1)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_columns:
    df1[col].replace({"Yes": 1, "No" : 0}, inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
for col in df1:
    print(f"{col} : {df1[col].unique()}")
    
```

```python
df1.gender.replace({"Female": 1, "Male": 0}, inplace=True)
df1.gender.value_counts()
```

```python
df1.gender.unique()
```

```python
df2 = pd.get_dummies(data=df1, columns=["InternetService", "Contract", "PaymentMethod"])
df2.head().T
```

```python
len(df2.columns), len(df1.columns)
```

```python
df2.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

col_to_scale = ["tenure", "MonthlyCharges", "TotalCharges"]
df2[col_to_scale] = scaler.fit_transform(df2[col_to_scale])
```

```python
df2.head().T
```

```python
df2.tenure.max(), df2.tenure.min()
```

```python
df2.MonthlyCharges.max(), df2.MonthlyCharges.min()
```

```python
df2.TotalCharges.max(), df2.TotalCharges.min()
```

```python
X = df2.drop("Churn", axis="columns")
y = df2["Churn"]
```

```python
X.shape, y.shape
```

```python
X.head().T
```

```python
y.T
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.shape, X_test.shape
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation="relu"),
#     keras.layers.Dense(15, activation="relu"),
    keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam",
             loss="binary_crossentropy",
             metrics=["accuracy"])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_preds = []
for element in yp:
    if element > 0.5:
        y_preds.append(1)
    else:
        y_preds.append(0)
```

```python
y_preds[:10], y_test[:10]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_preds))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_preds)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth');
```

```python
bank_df.drop(["RowNumber", "CustomerId", "Surname"],axis="columns", inplace=True)
```

```python
bank_df.head().T
```

```python
bank_df.Gender.replace({"Female":0, "Male":1}, inplace=True)
bank_df.head().T
```

```python
bank_df.isnull().sum()
```

```python
col_to_scale = ["CreditScore", "Age", "Tenure", "Balance", "EstimatedSalary", "NumOfProducts"]

scaler2 = MinMaxScaler()

bank_df[col_to_scale] = scaler2.fit_transform(bank_df[col_to_scale])
bank_df.head().T
```

```python
for col in bank_df:
    print(f"{col}: {bank_df[col].unique()}")
```

```python
bank_df.Geography.dtypes
```

```python
bank_df.Geography.unique()
```

```python
bank_df_2 = pd.get_dummies(data=bank_df, columns=["Geography"])
bank_df_2.head().T
```

```python
bank_df_2.dtypes
```

```python
bank_df_2.shape
```

```python
X = bank_df_2.drop("Exited", axis="columns")
y = bank_df_2["Exited"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=46)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(12, input_shape=(12,), activation="relu"),
    keras.layers.Dense(10, activation="relu"),
    keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam",
             loss="binary_crossentropy",
             metrics=["accuracy"])

model.fit(X_train, y_train, epochs=100)
```

```python
y_preds = model.predict(X_test)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_preds
```

```python
y_preds_2 = []
for element in y_preds:
    if element > 0.5:
        y_preds_2.append(1)
    else:
        y_preds_2.append(0)
```

```python
print(classification_report(y_test, y_preds_2))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_preds_2)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth');
```
